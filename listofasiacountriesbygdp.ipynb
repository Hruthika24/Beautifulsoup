```python
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as ply
import pandas as pd
```

```python
gdresponse=requests.get("https://en.wikipedia.org/wiki/List_of_Asian_countries_by_GDP")
gdresponse_text=gdresponse.text
type(gdresponse)
```

```python
gdsoup=BeautifulSoup(gdresponse_text,"html.parser")
```

```python
gdtables=gdsoup.find_all("table")
len(gdtables)
```

```python
heading_tags=gdtables[0].find_all("th")
heading_tags
```

```python
gdheadings=[i.text.strip() for i in heading_tags]
gdheadings
```

```python
row_tags=gdtables[0].find_all("td")
len(row_tags)
```

```python
rows=[i.text.strip() for i in row_tags]
rows
```

```python
rows=[i.text.strip() for i in row_tags]
rows[3:]
```

```python
gdrows=rows[3:]
gdrows
```

```python
gddict={}
for i in gdheadings:
    print(i)
```

```python
gddict={}
for i in gdheadings:
    gddict[i]=None
gddict
```

```python
gddict={}
n=0
for i in gdheadings:
    gddict[i]=[gdrows[j] for j in range(n,len(gdrows),len(gdheadings))]
    n+=1
gddict
```

```python
gddf=pd.DataFrame(gddict)
gddf.index=range(1,len(gddf)+1)
gddf
```

```python
gddf.rename(columns={"Country/Territory":"Country","2024 GDP(nominal) in billions":"GDP(billion)"},inplace=True)
gddf
```

```python
gddf.to_csv("gd")
```
